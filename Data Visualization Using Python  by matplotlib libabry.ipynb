```python
import matplotlib.patches as mpatches
```

```python

x1_axis = [2, 4, 6, 8]                          #legend
y1_axis = [1, 10, 100, 1000]
x2_axis = [1, 3, 5, 7]
y2_axis = [100, 110, 120, 130]

plt.title('My Third Graph')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend('exeponential', 'linear')

red_patch = mpatches.Patch(color = 'red', label = 'exeponentia')
green_patch = mpatches.Patch(color = 'green', label = 'linear')
plt.legend(handles = [green_patch, red_patch])
#plt.legend(handles = [red_patch])



plt.plot(x1_axis, y1_axis, 'r')
plt.plot(x2_axis, y2_axis, 'g')
plt.show()
```
